```python
Vaccine_canadamap_df = pd.read_csv('C:/Users/panka/OneDrive/Desktop/Class Folder/Github/project_3_for_COVID19/Datasets_ETLwork/Vaccine_record_Government_of_Canada/vaccination-coverage-map.csv',delimiter=',', header=0, na_values='NaN')
Vaccine_canadamap_df['Date'] = pd.to_datetime(Vaccine_canadamap_df['Date'])
Vaccine_canadamap_df.info()
```

```python
# Remove NaN values from specific columns
# remove NaN values and other common missing value representations from the 'column_name' column
missing_values = ['na', 'N/A', 'null', '']

# Replace missing value representations with NaN
Vaccine_canadamap_df['Cumulative number of people (Vaccinedose1)'] = Vaccine_canadamap_df['Cumulative number of people (Vaccinedose1)'].replace(missing_values, float('nan'))
Vaccine_canadamap_df['Cumulative percent of people (Vaccinedose1)'] = Vaccine_canadamap_df['Cumulative percent of people (Vaccinedose1)'].replace(missing_values, float('nan'))
# Drop NaN values from the column
Vaccine_canadamap_df.dropna(subset=['Cumulative number of people (Vaccinedose1)'], inplace=True)
Vaccine_canadamap_df.dropna(subset=['Cumulative percent of people (Vaccinedose1)'], inplace=True)

# convert the 'Cummulative number of people (Vaccinedose1)' column to int64
Vaccine_canadamap_df['Cumulative number of people (Vaccinedose1)'] = Vaccine_canadamap_df['Cumulative number of people (Vaccinedose1)'].astype('int64')
Vaccine_canadamap_df['Cumulative percent of people (Vaccinedose1)'] = Vaccine_canadamap_df['Cumulative percent of people (Vaccinedose1)'].astype('int64')
Vaccine_canadamap_df
```

```python
# Specify the entry you want to remove
entry_to_remove = 'Canada'

# Remove all rows with the specified entry in the 'Name' column
Vaccine_canadamap_df = Vaccine_canadamap_df[Vaccine_canadamap_df['Province'] != entry_to_remove]

# Display the modified dataframe
Vaccine_canadamap_df
```
